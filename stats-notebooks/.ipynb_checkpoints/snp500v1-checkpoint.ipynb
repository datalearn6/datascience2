```python
# useful functions
# ================
def total_return(prices):
    """Retuns the return between the first and last value of the DataFrame.

    Parameters
    ----------
    prices : pandas.Series or pandas.DataFrame

    Returns
    -------
    total_return : float or pandas.Series
        Depending on the input passed returns a float or a pandas.Series.
    """
    return prices.iloc[-1] / prices.iloc[0] - 1


def total_return_from_returns(returns):
    """Retuns the return between the first and last value of the DataFrame.

    Parameters
    ----------
    returns : pandas.Series or pandas.DataFrame

    Returns
    -------
    total_return : float or pandas.Series
        Depending on the input passed returns a float or a pandas.Series.
    """
    return (returns + 1).prod() - 1

def multi_period_return(period_returns):
     return np.prod(period_returns + 1) - 1

def plot_this(df, title, figsize=None, ylabel='',
             output_file='imgs/fig.png', bottom_adj=0.25,
             txt_ymin=-0.4, bar=False):
    if bar:
        ax = df.plot.bar(title=title, figsize=figsize)
    else:
        ax = df.plot(title=title, figsize=figsize)
    sns.despine()
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.text(0, txt_ymin, asset_info, transform=ax.transAxes, fontsize=9)
    plt.gcf().subplots_adjust(bottom=bottom_adj)
    plt.savefig(output_file, **kw_save)
```

```python
## SP500 contains the monthly datasets for S&P figures

#sp_data = pd.read_csv('sp500.csv')
sp_data = pd.read_csv('./data/sp500.csv',
                      header=0,
                      names=['timepoint','open','high','low','close','adjclose','volume'],
                      parse_dates=True,index_col=0)


savings_data = pd.read_csv('./data/savings_interest.csv',
                      header=0,
                      names=['timepoint','rate'],
                      parse_dates=True,index_col=0)


savings_data = savings_data[datetime.date(year=1990,month=1,day=1):datetime.date(year=2017,month=4,day=1)]
```

```python
## Summary of S&P Daily dataset (in points)

print(sp_data.columns)

sp_data.describe()
```

```python
## Summary of Daily Interest Rates dataset (in percentages)

##Convert the data from percentages to actual returns
savings_data = savings_data.mul(0.01)


print(savings_data.columns)
savings_data.describe()
```

```python

## Assumptions that GIC would be .0125 (or 1.25% more than savings account interest)

offset_gic_savings = 0.0125

gic_data = savings_data.apply(lambda x:x+.0125)

#gic_data = savings_data(lambda row:row['rate']+.0125)

print(gic_data.columns)
gic_data.describe()
```

```python
## Filling the frequency of continuous days
sp_data = sp_data.asfreq(freq='1D', method='ffill')

savings_data = savings_data.asfreq(freq='1D', method='ffill')

gic_data = gic_data.asfreq(freq='1D', method='ffill')
```

```python
## Calculating S&P Return Dataset

##Assumption:
##Returns calculated daily.(approach 1)
##Returns calculated on monthly average means.(approach 2)
sp_approach1 = sp_data['adjclose'].pct_change()
rolling_return_1 = sp_approach1.rolling('1095D').apply(multi_period_return).fillna(0)



sp_plot_data = [go.Scatter(x=sp_data.index, y=sp_data.adjclose, name='S&P 500 Index'),
               go.Scatter(x=sp_approach1.index, y=sp_approach1.adjclose, name='S&P 500 Daily Return'),
               go.Scatter(x=rolling_return_1.index, y=rolling_return_1.adjclose, name='S&P 500 Rolling Return')
               ]

iplot(sp_plot_data,filename='sp_plot_data')
```

```python
rates_return_data = [go.Scatter(x=savings_data.index, y=savings_data.rate, name='Savings Rate'),
               go.Scatter(x=gic_data.index, y=gic_data.rate, name='GIC Rate')]

iplot(rates_return_data)
```
