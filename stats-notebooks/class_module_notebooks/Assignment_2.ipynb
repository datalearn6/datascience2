```python
import random
import pandas as pd
import numpy as np

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
def plot_probs(x_, y_):
    trace = go.Scatter(
        x = x_,
        y = y_
    )

    layout = go.Layout(
        xaxis=dict(
            title='Number of Simulations'
        ),
        yaxis=dict(
            title='Probability'
        )
    )

    fig = go.Figure(data = [trace], layout = layout)
    return fig
```

Anwer 1 : Given you are parent of twins(both are girls), find out the probability that they are identical ?

```python
are_identical_twins=0.3
## Assuming the total probability is 1, the other set possible is fraternal whose probability = (1-.3)
not_identical_twins=0.7

are_identical_twins_girls=.5
are_identical_twins_boys=.5

not_identical_twins_boys=.25
not_identical_twins_girls=.25
not_identical_oneboy_one_girl=.5

## Given parents of twins and both are girls, what's the probability that they are identical ?
prob_twins_girls_are_identical = are_identical_twins * are_identical_twins_girls
print("Probability of twin girls is %.3f if they are identical"  %prob_twins_girls_are_identical)

prob_twins_girls_not_identical = not_identical_twins * not_identical_twins_girls
print("Probability of twin girls is %.3f if they are not identical" % prob_twins_girls_not_identical)

identical_given_twin_girls = prob_twins_girls_are_identical / (prob_twins_girls_are_identical + prob_twins_girls_not_identical)

output_percent = identical_given_twin_girls*100
print("If you have a twin girls, there is a %.3f%% chance that they would be identical" % output_percent)
```

```python
twin_scenarios = np.arange(1,6)
effective_simulations = 0
count_hits = 0
x_data = []
y_data = [] 
 
for num_simulations in range(1, 50000):
    twins = np.random.choice(twin_scenarios,p=[0.15,  0.15,  0.175,  0.175,  0.35])
    if twins==2 or twins == 3:
        effective_simulations += 1
        if twins == 2:
            count_hits += 1
        if (effective_simulations % 10) == 0:
            y_data.append(float(count_hits) / float(effective_simulations))
            x_data.append(effective_simulations)

# Plot the figure
fig = plot_probs(x_data, y_data)
iplot(fig)
```

```python
## Answer 3
## Porbability model when winning x dollars has different values
##when x = 5
odds_of_spade=13.0/52.0

## when x=10
odds_of_club_regular=(12.0/52.0)

## when x=30
odds_of_club_ace=(1.0/52.0)

## when x=0
odds_of_losing = 26.0/52.0

#a) How much can you expect to win:
expected_win_value = odds_of_spade * 5 + odds_of_club_regular * 10 + odds_of_club_ace *30
print("You can expect to win $%.2f" % expected_win_value)

variance_in_win_value = ((5 - expected_win_value)**2 * odds_of_spade) + \
                        ((10.0 - expected_win_value)**2 * odds_of_club_regular) + \
                        ((30 - expected_win_value)**2 * odds_of_club_ace) + \
                        ((0 - expected_win_value)**2 * odds_of_losing)

std_in_win_value = np.sqrt(variance_in_win_value)

print("The variance in the win value is $%.2f" % variance_in_win_value)
print("The standard deviation in the win value is $%.2f" % std_in_win_value)
```
